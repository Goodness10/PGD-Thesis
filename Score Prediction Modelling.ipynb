# Modelling

```python
#splitting into dependent (y) and independent (X) variables
X = df.drop("score" , axis = 1)
y = df["score"]

X.shape
```

```python
# Get the categorical features
d_types = df.dtypes

# Filter the categorical columns
cat_cols = d_types[d_types == 'object'].index.tolist()

cat_cols
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Assuming you have your feature matrix X with both categorical and numerical features
categorical_features = cat_cols  # List of categorical feature names
numerical_features = ['age', 'weeklytime']  # List of numerical feature names

# Separate the data into categorical and numerical features
X_categorical = X[categorical_features]
X_numerical = X[numerical_features]

# Create a OneHotEncoder for encoding categorical features
cat_encoder = OneHotEncoder(handle_unknown='ignore')

# Fit and transform the OneHotEncoder to the categorical features
X_categorical_encoded = cat_encoder.fit_transform(X_categorical)

# Get the final column names after one-hot encoding
cat_columns = cat_encoder.get_feature_names(input_features=categorical_features)

# Convert the encoded categorical features back to a DataFrame with the appropriate column names
X_categorical_encoded_df = pd.DataFrame(X_categorical_encoded.toarray(), columns=cat_columns)

# Create a StandardScaler for scaling numerical features
scaler = StandardScaler()

# Fit and transform the StandardScaler to the numerical features
X_numerical_scaled = scaler.fit_transform(X_numerical)

# Convert the scaled numerical features back to a DataFrame with the appropriate column names
X_numerical_scaled_df = pd.DataFrame(X_numerical_scaled, columns=numerical_features)

# Concatenate the encoded categorical features and scaled numerical features into a single DataFrame
X_encoded_df = pd.concat([X_categorical_encoded_df, X_numerical_scaled_df], axis=1)

#print(X_encoded_df)
```

# With Other Models

```python
from sklearn.model_selection import KFold
from catboost import CatBoostRegressor
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
import lightgbm as lgb
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor


# Create an empty dataframe to store the results
results_df = pd.DataFrame(columns=['Model', 'RMSE', 'MAE', 'MSE'])

def result(model, splits, model_name):
    # Initialize KFold with the desired number of folds
    kf = KFold(n_splits=splits, shuffle=True, random_state=42)

    # Create empty lists to store the evaluation metrics for each fold
    rmse_scores = []
    mae_scores = []
    mse_scores = []
   
    # Iterate through the KFold splits
    for train_index, test_index in kf.split(X_encoded_df):
        # Split the data into train and test sets based on the indices provided by KFold
        X_train, X_test = X_encoded_df.iloc[train_index], X_encoded_df.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        # Fit the model to the training data
        model.fit(X_train, y_train)

        # Make predictions on the test data
        y_pred = model.predict(X_test)

        # Calculate evaluation metrics
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        mae = mean_absolute_error(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        

        # Append the metrics to the corresponding lists
        rmse_scores.append(rmse)
        mae_scores.append(mae)
        mse_scores.append(mse)
        

    # Calculate the average of the evaluation metrics
    avg_rmse = np.mean(rmse_scores).round(3)
    avg_mae = np.mean(mae_scores).round(3)
    avg_mse = np.mean(mse_scores).round(3)
    

    # Append the results to the dataframe
    results_df.loc[len(results_df)] = [model_name, avg_rmse, avg_mae, avg_mse]
```

```python
# Run the models and save the results for 5 splits
result(GradientBoostingRegressor(random_state = 42), 5, "Gradient Boosting5")
result(CatBoostRegressor(random_state = 42, silent = True), 5, "CatBoost5")
result(lgb.LGBMRegressor(random_state = 42), 5, "LightGBM5")
result(RandomForestRegressor(random_state = 42), 5, "Random Forest5")
result(SVR(), 5, "Support Vector Regressor5")

# Print the results dataframe
print(results_df)
```

# Support Vector Machine

```python
kf = KFold(n_splits = 6, shuffle=True, random_state=42)

# Create empty lists to store the evaluation metrics for each fold
rmse_scores = []
mae_scores = []
r2_scores = []
mse_scores = []

    # Iterate through the KFold splits
for train_index, test_index in kf.split(X_encoded_df):
    # Split the data into train and test sets based on the indices provided by KFold
    X_train, X_test = X_encoded_df.iloc[train_index], X_encoded_df.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Fit the model to the training data
    model_svr = SVR(kernel = "linear")
    model_svr.fit(X_train, y_train)

    # Make predictions on the test data
    y_pred = model_svr.predict(X_test)

    # Calculate evaluation metrics
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
   

    # Append the metrics to the corresponding lists
    rmse_scores.append(rmse)
    mae_scores.append(mae)
    r2_scores.append(r2)
    mse_scores.append(mse)
    
# Calculate the average of the evaluation metrics
avg_rmse = np.mean(rmse_scores).round(3)
avg_mae = np.mean(mae_scores).round(3)
avg_r2 = np.mean(r2_scores).round(3)
avg_mse = np.mean(mse_scores).round(3)
print(avg_rmse)
print(avg_mae)
print(avg_r2)
print(avg_mse)
```

```python
coefficients = model_svr.coef_[0]

# Create a DataFrame to store feature importances
feature_importance_df = pd.DataFrame({'Feature': X_encoded_df.columns, 'Coefficient': coefficients})

# Extract the original column names from the feature names
feature_importance_df['Original_Column'] = feature_importance_df['Feature'].str.split('_').str[0]

# Aggregate coefficient scores by summing them for each original column
grouped_importance = feature_importance_df.groupby('Original_Column')['Coefficient'].sum().reset_index()

# Sort the DataFrame by importance scores in descending order
grouped_importance = grouped_importance.sort_values('Coefficient', ascending=False)

# Print the sorted feature importance
print(grouped_importance)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(grouped_importance['Original_Column'], grouped_importance['Coefficient'])
plt.xticks(rotation=90, ha='center', size = 15)
plt.xlabel('Features', size = 15)
plt.ylabel('Importance Score', size = 15)
plt.title('Grouped Feature Importance for Random Forest', size = 20)
plt.tight_layout()
plt.savefig("Grouped Feature Importance for Random Forest.png")
plt.show()
```
